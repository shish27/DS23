```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Задание 1
Постройте график экспоненты. Подпишите оси. Дайте название графику.

```python
x = np.linspace(0, 5, 100)
y = np.exp(x)
plt.plot(x, y, label='y = e^x', color='red')
plt.grid()
plt.title('График экспоненты')
plt.xlabel('Ось  x ')
plt.ylabel('Ось  y ')
plt.legend()
plt.show()
```

Задание 2
Постройте график функции $\rho = 4 \sin(2\phi)$ в полярных координатах. Линия должна быть любого нестандартного цвета из доступных по ссылке ниже (на ваш вкус). Дайте название графику.
https://matplotlib.org/3.1.0/gallery/color/named_colors.html

```python
plt.figure(1, figsize=(10,7))
r = np.arange(0, 2, 0.01)
q = 2 * np.pi * r
y = 2* np.sin(2*q)
ax = plt.subplot(projection='polar')
ax.plot(y, c= 'darkgreen')
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2]) 
plt.title('  ρ=4sin(2ϕ) ')
plt.show()
```

Задание 3

Постройте графики функций $y = x^3$ и $y = x^2$ на одном графике. 
Дайте название графику;
Дайте названия осям;
Присвойте лейблы и выведите легенду;
Графики функций должены быть отрисованы пунктирными линиями (любыми из доступных на ваш вкус)

```python
plt.figure(1, figsize=(8,5))
x = np.linspace(0, 10, 100)
y1 = x**3
y2 =x**2
plt.plot(x, y1, label = 'y = x^3', color = 'blue', ls='--')
plt.plot(x, y2, label = 'y = x^2', color = 'fuchsia', ls='--')
plt.grid()
plt.title('График функций y = x^3 и y = x^2')
plt.xlabel('Ось  x ')
plt.ylabel('Ось  y ')
plt.legend()
plt.show()
```
