```python
iris_df = DataFrame(data.data)
```

```python
iris_df.columns = data.feature_names
```

```python
iris_df['target'] = data.target
```

```python
iris_df['target_names'] = iris_df.target.apply(lambda x : data.target_names[x])
```

```python
plt.figure(figsize = (8, 5))
sns.pairplot(data = iris_df, hue= 'target')
```

```python
plt.figure(figsize = (8, 7))
sns.boxplot(data = iris_df, palette = 'Set2')
```

```python
plt.figure(figsize = (10,6))
sns.violinplot(data = iris_df, palette = 'Set2')
```

```python
plt.subplots(figsize = (16,20))

plot_number = 0
for feature_name in data['feature_names']:
  for target_name in data['target_names']:
    plot_number += 1
    plt.subplot(4,3, plot_number)
    sns.boxplot(iris_df[iris_df.target == target][feature_name])
    plt.title(target_name)
    plt.xlabel('cm')
    plt.ylabel(feature_name[:-4])
```
